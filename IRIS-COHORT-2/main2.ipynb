```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
import pandas as pd

# Assuming data is loaded from a CSV file
# Replace this with the actual loading method if you're reading from a file
data = pd.read_csv(r'C:\Users\OMOLP049\Documents\MitaliRao\Data - Sheet1.csv')  # Replace with the actual file path

# Check the first few rows to confirm the dataset is loaded
print(data.head())
```

# Graph 1

```python
## Graph 1: Focus Score vs. Music Genre
df = pd.DataFrame(data)

# Graph 1: Focus Score vs. Music Genre
plt.figure(figsize=(10, 6))
plt.bar(df['Music Genre'], df['Focus %'], color=['blue', 'green', 'orange', 'red'])
plt.title("Focus Score Across Music Genres", fontsize=14)
plt.xlabel("Music Genre", fontsize=12)
plt.ylabel("Focus Score", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Ensure 'Focus %' is a string if it contains '%'
if data['Focus %'].dtype == 'object':
    data['Focus %'] = data['Focus %'].str.rstrip('%').astype(float)
else:
    data['Focus %'] = data['Focus %'].astype(float)

# Plot Focus % by Music Genre
plt.figure(figsize=(10, 6))
sns.barplot(x='Music Genre', y='Focus %', data=data, palette='muted')
plt.title('Effect of Music Genre on Concentration (Focus %)', fontsize=14)
plt.ylabel('Focus (%)')
plt.xlabel('Music Genre')
plt.show()
```

# Graph 2

```python
# Graph 2: Eye Movement Variability (Fixation Duration) Standard Deviation vs Music Genre
fixation_std = df['Fixation Duration (ms)'].std()

plt.figure(figsize=(10, 6))
plt.bar(df['Music Genre'], df['Fixation Duration (ms)'], color=['purple', 'cyan', 'magenta', 'yellow'])
plt.axhline(y=fixation_std, color='black', linestyle='--', label=f'Standard Deviation = {fixation_std:.2f}')
plt.title("Fixation Duration Across Music Genres", fontsize=14)
plt.xlabel("Music Genre", fontsize=12)
plt.ylabel("Fixation Duration (ms)", fontsize=12)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Ensure 'Confusion %' is in a usable format
if data['Confusion %'].dtype == 'object':
    data['Confusion %'] = data['Confusion %'].str.rstrip('%').astype(float)
else:
    data['Confusion %'] = data['Confusion %'].astype(float)

# Scatter plot of SD vs. Confusion %
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['Fixation_SD'], y=data['Confusion %'], label='Fixation Duration SD', color='blue')
sns.scatterplot(x=data['Saccadic_Velocity_SD'], y=data['Confusion %'], label='Saccadic Velocity SD', color='orange')
plt.title('Reliability of Eye Movement Variability as an Indicator of Cognitive Distraction', fontsize=14)
plt.xlabel('Standard Deviation of Eye Movements')
plt.ylabel('Confusion (%)')
plt.legend()
plt.show()
```

# Graph 3

```python
# Graph 3: Correlation Between Saccadic Velocity and Focus Score
plt.figure(figsize=(10, 6))
plt.scatter(df['Saccadic Velocity (°/s)'], df['Focus %'], color='blue', edgecolor='k')
plt.title("Correlation Between Saccadic Velocity and Focus Score", fontsize=14)
plt.xlabel("Saccadic Velocity (°/s)", fontsize=12)
plt.ylabel("Focus Score", fontsize=12)
plt.grid(linestyle='--', alpha=0.7)
plt.show()
```

```python
# Scatter plot for Fixation Count vs. Focus %
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Fixation Count', y='Focus %', data=data, color='blue')
plt.title('Correlation Between Eye Movement and Focus', fontsize=14)
plt.xlabel('Fixation Count')
plt.ylabel('Focus (%)')
plt.show()
```

# Graph 4

```python
# Graph 4: Pupil Dilation Across Music Genres
plt.figure(figsize=(10, 6))
plt.plot(df['Music Genre'], df['Pupil Dilation (mm)'], marker='o', color='green', label="Pupil Dilation")
plt.title("Pupil Dilation Across Music Genres", fontsize=14)
plt.xlabel("Music Genre", fontsize=12)
plt.ylabel("Pupil Dilation (mm)", fontsize=12)
plt.grid(linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
# Plot Focus Score by Music Genre
plt.figure(figsize=(10, 6))
sns.boxplot(x='Music Genre', y='Focus Score', data=data, palette='coolwarm')
plt.title('Effect of Combining Biometric and Auditory Data on Focus Score', fontsize=14)
plt.ylabel('Focus Score')
plt.xlabel('Music Genre')
plt.show()
```

# Graph 5

```python
# Graph 5: Variability in Facial Expression Dynamics
facial_dynamics_means = [np.mean(x) for x in df['Facial Expression Dynamics']]

plt.figure(figsize=(10, 6))
plt.bar(df['Music Genre'], facial_dynamics_means, color=['red', 'blue', 'green', 'purple'])
plt.title("Mean Facial Expression Dynamics Across Music Genres", fontsize=14)
plt.xlabel("Music Genre", fontsize=12)
plt.ylabel("Mean Facial Expression Dynamics", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Ensure 'Confusion %' is numeric
if data['Confusion %'].dtype == 'object':
    data['Confusion %'] = data['Confusion %'].str.rstrip('%').astype(float)
else:
    data['Confusion %'] = data['Confusion %'].astype(float)

# Scatter plot for Focus % vs. Confusion %
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Focus %', y='Confusion %', data=data, color='purple')
plt.title('Objective Data (Focus %) vs. Subjective Data (Confusion %)', fontsize=14)
plt.xlabel('Focus (%)')
plt.ylabel('Confusion (%)')
plt.show()
```
